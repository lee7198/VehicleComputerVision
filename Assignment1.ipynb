```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/gdrive')
```

## RGB 이미지를 구해서(편한 방법으로 구하면 됩니다) 아래의 작업을 수행하시오.

```python
img = cv2.imread('/content/gdrive/My Drive/차량비전시스템/src/campus.jpeg')
```

### 1. RGB 이미지 출력

```python
cv2_imshow(img)
```

```python
RGB_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # BGR을 RGB로 변환
Red_img, Green_img, Blue_img = cv2.split(RGB_img) #RGB 채널 분리
```

### 2. R-channel 출력

```python
cv2_imshow(Red_img)
```

### 3. G-channel 출력

```python
cv2_imshow(Green_img)
```

### 4. B-channel 출력

```python
cv2_imshow(Blue_img)
```

### 5. Gray scale 이미지로 변환

```python
Gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY) #gray 스케일로 변환
cv2_imshow(Gray_img)
```

### 6. Binary 이미지로 변환

```python
# 100을 임계값으로 선언
ret, dst = cv2.threshold(Gray_img, 100, 255, cv2.THRESH_BINARY)
cv2_imshow(dst)
```

```python
# 2차원 배열을 1차원 배열로 변환
d1Arr = np.concatenate(dst)
```

### 7. 5)번 작업을 수행한 후 출력 이미지에서 최대 화소값 출력

```python
print('max: ', max(d1Arr))
```

### 8. 5)번 작업을 수행한 후 출력 이미지에서 최소 화소값 출력

```python
print('min: ', min(d1Arr))
```

### 9. 5)번 작업을 수행한 후 전체 화소값의 평균값 출력


```python
print('avg: ', np.mean(d1Arr))
```

## RGB 이미지 두 개(두 개의 이미지 사이즈가 동일해야 함)를 구해서 아래의 작업을 수행하시오



```python
#chunsik = cv2.imread('/content/gdrive/My Drive/차량비전시스템/src/춘무룩.jpeg')
moon = cv2.imread('/content/gdrive/My Drive/차량비전시스템/src/night.jpeg')
campus = cv2.imread('/content/gdrive/My Drive/차량비전시스템/src/공대.jpeg')
cv2_imshow(campus)
```

### 1. 두 이미지 모두 Gray scale로 변환

```python
Gray_campus = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
Gray_campus2 = cv2.cvtColor(campus, cv2.COLOR_BGR2GRAY)
d1Arr = np.concatenate(Gray_campus)
print('max: ', max(d1Arr))
d1Arr = np.concatenate(Gray_campus2)
print('max: ', max(d1Arr))
```

### 2. 두 이미지의 화소값을 모두 더하시오

#### a) Saturation 방법을 사용할 경우의 결과를 출력하시오

```python
img_plus = Gray_campus + Gray_campus2
cv2_imshow(img_plus)
```

#### b) Wrapping 방법을 사용할 경우의 결과를 출력하시오

```python
img_add = cv2.add(Gray_campus, Gray_campus2)
cv2_imshow(img_add)
```
