## 3. flower.bmp 이용하여 다음을 수행하고 결과 비교 분석

```python
img = cv2.imread('/content/drive/MyDrive/차량비전시스템/src/asg3/flower.bmp')
gray_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

### sobel 적용 후 엣지 검출

```python
# sobel operator
x = cv2.Sobel(gray_img, -1, 1, 0)
y = cv2.Sobel(gray_img, -1, 0, 1)
#uint8 타입(0~255)로 변경하고 영상 합치기
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
output_img = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)

plt.figure(figsize=(20,20))
plt.subplot(1, 4, 1)
plt.title('original image')
plt.imshow(gray_img, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 2)
plt.title('sobel edge detection')
plt.imshow(output_img, cmap='gray')
plt.axis("off")
```

### prewitt 적용 후 엣지 검출

```python
# prewitt filter
gx = np.array([[-1,0,1], [-1,0,1], [-1,0,1]], dtype=int)
gy = np.array([[1,1,1], [0,0,0], [-1,-1,-1]], dtype=int)
# prewitt filter convolution
x = cv2.filter2D(gray_img, -1, gx)
y = cv2.filter2D(gray_img, -1, gy)
#uint8 타입(0~255)로 변경하고 영상 합치기
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
output_img = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)

plt.figure(figsize=(20,20))
plt.subplot(1, 4, 1)
plt.title('original image')
plt.imshow(gray_img, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 2)
plt.title('prewitt edge detection')
plt.imshow(output_img, cmap='gray')
plt.axis("off")
```

### canny 적용 후 엣지 검출

```python
# canny edge convolution, T1=100, T2=200
output_img = cv2.Canny(gray_img, 100, 200)

plt.figure(figsize=(20,20))
plt.subplot(1, 4, 1)
plt.title('original image')
plt.imshow(gray_img, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 2)
plt.title('canny edge detection T1=100, T2=200')
plt.imshow(output_img, cmap='gray')
plt.axis("off")

output_img = cv2.Canny(gray_img, 250, 200)
plt.subplot(1, 4, 3)
plt.title('canny edge detection T1=100, T2=200')
plt.imshow(output_img, cmap='gray')
plt.axis("off")
```

### robert cross 적용 후 엣지 검출

```python
# robert cross filter
gx = np.array([[-1,0], [0,1]], dtype=int)
gy = np.array([[0,-1], [1,0]], dtype=int)
# robert cross convolution
x = cv2.filter2D(gray_img, -1, gx)
y = cv2.filter2D(gray_img, -1, gy)

#uint8 타입(0~255)로 변경하고 영상 합치기
absX = cv2.convertScaleAbs(x)
absY = cv2.convertScaleAbs(y)
output_img = cv2.addWeighted(absX, 0.5, absY, 0.5, 0)

plt.figure(figsize=(20,20))
plt.subplot(1, 4, 1)
plt.title('original image')
plt.imshow(gray_img, cmap='gray')
plt.axis("off")

plt.subplot(1, 4, 2)
plt.title('roberts cross edge detection')
plt.imshow(output_img, cmap='gray')
plt.axis("off")
```
